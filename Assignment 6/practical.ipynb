```python
# q1_combinations.py
import itertools


def all_combinations(d):
# Ensure deterministic order by sorting keys (optional)
    values = [d[k] for k in sorted(d.keys())]
    for combo in itertools.product(*values):
        print(''.join(combo))


if __name__ == '__main__':
    data = {'1': ['a', 'b'], '2': ['c', 'd']}
all_combinations(data)
```

```python
def top_n_items(d, n=3):
# sort items by value descending
    return sorted(d.items(), key=lambda x: x[1], reverse=True)[:n]


if __name__ == '__main__':
    shop_items = {'item1': 45.50, 'item2': 35, 'item3': 41.30, 'item4': 55, 'item5': 24}
for item, price in top_n_items(shop_items, 3):
    print(f"{item} {price}")
```

```python
# q3_max_values.py


def find_top_keys(d, n):
    
    sorted_items = sorted(d.items(), key=lambda kv: kv[1], reverse=True)
    return [k for k, v in sorted_items[:n]]


if __name__ == '__main__':
    data = {'a': 5, 'b': 14, 'c': 32, 'd': 35, 'e': 24, 'f': 100, 'g': 57, 'h': 8, 'i': 100}


print("1 maximum value(s):", find_top_keys(data, 1))
print("2 maximum value(s):", find_top_keys(data, 2))
print("5 maximum value(s):", find_top_keys(data, 5))
```

```python
# q4_combine_lists.py


def combine_lists(keys, values):
    return dict(zip(keys, values))


if __name__ == '__main__':
    keys = ['a', 'b', 'c', 'd', 'e', 'f']
    values = [1, 2, 3, 4, 5]
    combined = combine_lists(keys, values)
    print(combined)
```

```python
def evaluate(expr, variables=None):
    """
    Evaluate a mathematical expression represented as a dictionary.
    Supports +, -, *, / operations, variables, and nested expressions.
    """
    if variables is None:
        variables = {}

    
    if "value" in expr:
        return expr["value"]

    
    elif "var" in expr:
        var_name = expr["var"]
        if var_name not in variables:
            raise ValueError(f"Variable '{var_name}' is not defined.")
        return variables[var_name]

    
    elif "op" in expr:
        op = expr["op"]
        args = expr["args"]

        
        values = [evaluate(arg, variables) for arg in args]

        
        if op == "+":
            return sum(values)
        elif op == "-":
            return values[0] - sum(values[1:])
        elif op == "*":
            result = 1
            for val in values:
                result *= val
            return result
        elif op == "/":
            result = values[0]
            for val in values[1:]:
                if val == 0:
                    raise ZeroDivisionError("Division by zero!")
                result /= val
            return result
        else:
            raise ValueError(f"Unknown operator '{op}'")

    else:
        raise ValueError("Invalid expression format.")



expr1 = {
    "op": "*",
    "args": [
        {"op": "+", "args": [{"value": 3}, {"value": 5}]},
        {"value": 2}
    ]
}


expr2 = {
    "op": "*",
    "args": [
        {"op": "+", "args": [{"var": "x"}, {"value": 2}]},
        {"var": "y"}
    ]
}


expr3 = {
    "op": "+",
    "args": [
        {"op": "/", "args": [
            {"op": "-", "args": [{"value": 10}, {"value": 4}]},
            {"value": 3}
        ]},
        {"op": "*", "args": [{"value": 2}, {"value": 5}]}
    ]
}



print("Expression 1:", evaluate(expr1))                          
print("Expression 2:", evaluate(expr2, {"x": 4, "y": 3}))        
print("Expression 3:", evaluate(expr3))                          
```
